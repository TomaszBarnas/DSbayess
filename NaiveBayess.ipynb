```python
import numpy as np
import random as rd
import pandas as pd

iris = pd.read_csv("iris.csv")

class ProcessingData:
    @staticmethod
    def shuffle(x):
        #x.iloc[i]
        #x.at[x, j]
        for i in range(len(x)-1,0,-1):
            #random miejsce
            j = rd.randint(0,i)
            #zamiania rekordow
            x.iloc[i], x.iloc[j] = x.iloc[j],x.iloc[i]
        return x
    @staticmethod
    def split(x, k): #k=0.7
        n = int(len(x) * k)
        xTrain = x[:n]
        xVal = x[n:]
        
        return xTrain, xVal

class Bayes:
    @staticmethod
    def probability(sample):
        classProbability = []

        # KAŻDA KLASA
        for currentClassIndex in range(3):
            columnProbability = []

            # KAŻDA KOLUMNA
            for currentColumnIndex in range(4):
                currentValue = sample[currentColumnIndex]

                # PRAWDOPODOBIEŃSTWO WARUNKOWE KOLUMNY
                columnProbability.append((Bayes.normal(currentValue, means.iloc[currentClassIndex][currentColumnIndex], std.iloc[currentClassIndex][currentColumnIndex])))

            # WYMNOŻONE WSZYSTKIE PRAWDOPODOBIEŃSTWA DLA KLASY
            classProbability.append(prior.iloc[currentClassIndex]*columnProbability[0]*columnProbability[1]*columnProbability[2]*columnProbability[3])

        # PORÓWNANIE WYNIKÓW
        # print("Oczekiwane: ", sample["variety"], ", otrzymane: ", classes[classProbability.index(max(classProbability))])

        return classes[classProbability.index(max(classProbability))]
        
    @staticmethod
    def normal(data, mean, std):
        pdf = (np.e ** -((data - mean)**2) / (2*(std)**2)) / (std * np.sqrt(2 * np.pi))
        return pdf
    

iris1 = ProcessingData.shuffle(iris)
#iris1.head

irisTrain, irisVal = ProcessingData.split(iris1, 0.7)

means = irisTrain.groupby(["variety"]).mean()   # ŚREDNIA
std = irisTrain.groupby(["variety"]).std()   # ODCHYLENIE STANDARDOWE
prior = (irisTrain.groupby("variety").count() / len(irisTrain)).iloc[:,1]   # PRAWDOPODOBIEŃSTWO KAŻDEJ Z KLAS
classes = np.unique(irisTrain["variety"].tolist())   # WSZYSTKIE KLASY

corrected = 0
for rowIndex in range(len(irisVal)):
    if irisVal.iloc[rowIndex]["variety"] == Bayes.probability(irisVal.iloc[rowIndex]):
        corrected += 1

print("Dokladnosc naiwnego klasyfikatora Bayesa to: ", (corrected/len(irisVal)*100))

means

std

prior

classes
```
